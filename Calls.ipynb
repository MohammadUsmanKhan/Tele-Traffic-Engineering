```python
"""Name : Muhammad Usman Khan
   Roll no. : 2017-TE-58
   Section : B
   Subject : Tele-Traffic Engineering
"""

import matplotlib.pyplot as plt #importing the library for ploting

#Takinh Run time input values of LAMDA(t)
Lamdat=[]
for i in range(0,7):
    c=int(input("Enter "+str(i)+" of Lamda(t) :"))
    Lamdat.append(c)
print("\n"+"The Values of Lamda(t)={}".format(Lamdat))

#Taking Run time input values of Blocking Probability
B=[]
for i in range(0,7):
    c=float(input("Enter "+str(i)+" of B :"))
    B.append(c)
print("\n"+"The Value of Blocking Probabilities={}".format(B))

#Taking Run time input values of Probability of not answering
PNA=[]
for i in range(0,7):
    c=float(input("Enter "+str(i)+" of PNA :"))
    PNA.append(c)
print("\n"+"The Values of Probability of not answering={}".format(PNA))

#Taking Run time input values of Probability of repeat
PR=[]
for i in range(0,7):
    c=float(input("Enter "+str(i)+" of PR :"))
    PR.append(c)
print("\n"+"The Values of Probability of repeat={}".format(PR))

#Taking Run time input values of Percentages of users in an Hour
PER=[]    
for i in range(0,7):
    c=int(input(
    "Enter "+str(i)+" of PER :"))
    PER.append(c)
print("\n"+"The Values of percentage of users in an hour={}".format(PER))

#Taking Run time input values of Number of users in an Hour
no_of_users=[]
for i in range(0,7):
    c=int(input("Enter "+str(i)+" of no_of_users :"))
    no_of_users.append(c)
print("\n"+"The values of Number of users in an hour={}".format(no_of_users))
    
tm=[1,2,3,4,5,6,7] #Values of time in Hour

lamT=[] #initializing the empyty array for appending vales of LAMDA(T)
lamC=[] #initializing the empyty array for appending vales of LAMDA(C)
LamdaR=[] #initializing the empyty array for appending vales of LAMDA(R)

#Using loop for calculation and changing value index 
for a in range(0,7):
    a+=0 # Ilteration 
    limr= (Lamdat[a]*(B[a]+(1-B[a])*PNA[a])*PR[a]) #formula of LAMDA(R)
    
    LamdaC=((no_of_users[a])/100)*PER[a] #formula of LAMDA(C)
    
   
    LamdaR.append(limr) # appending values in empty array of LamdaR
    
    lamC.append(LamdaC) # appending values in empty array of LamdaC
    LamdaT=LamdaR[a]+Lamdat[a] #formula of LAMDA(T)
    lamT.append(LamdaT) #appending values in empty array of lamT
    
print(lamT) #printing values of LAMDA(T)
print(LamdaR) #printing values of LAMDA(R)
print(lamC) #printing values of LAMDA(C)

#ploting the Call Rpeat Lamda(R)
plt.subplot(3,1,1)
plt.plot(tm, LamdaR,"ob--",color='r')
plt.xlabel('Time (hr)')
plt.ylabel('call repeat')
plt.title("Call Rpeat Lamda(R)")

#ploting the Total Arrival Rate Lamda(T)
plt.subplot(3,1,2)
plt.plot(tm,lamT,"ob--",color='g')
plt.xlabel("Time(hr)")
plt.ylabel("Total Arrival Rate")
plt.title("Total Arrival Rate Lamda(T)")

#ploting the Users in per Hour Lamda(C)
plt.subplot(3,1,3)
plt.plot(tm,lamC,"ob--",color='orange')
plt.xlabel("Time(hr)")
plt.ylabel("Users in per Hour")
plt.title("Users in per Hour Lamda(C)")
```
